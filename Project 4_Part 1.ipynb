```python
# this cell can take up to 1 min to run
date_format = "%m/%d/%Y %H:%M:%S %p"

chicago_full['Datetime'] = pd.to_datetime(chicago_full['Date'], format=date_format)
chicago_full = chicago_full.sort_values('Datetime')
chicago_full
```

```python
## Dropping unneeded columns to reduce file size
drop_cols = ["X Coordinate","Y Coordinate", "Community Area","FBI Code",
             "Case Number","Updated On",'Block','Location','IUCR']
```

# How do crime rates differ between police districts?
Which district has the most crimes? Which has the least?

```python
district_crime_counts = df['District'].value_counts()
district_crime_counts
```

```python
district_crime_rates = district_crime_counts/df['District'].sum()
district_crime_rates
```

```python
highest_crime_district = district_crime_rates.idxmax()
highest_crime_district
```

```python
lowest_crime_district = district_crime_rates.idxmin()
lowest_crime_district
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
district_crime_counts.plot(kind='barh')
plt.title('Crime Rates in Different Districts')
plt.xlabel('Number of Crimes')
plt.ylabel('Districts')
plt.show()
```

```python
print(f'District with the highest crime rate: {highest_crime_district} with approximately more than 500,000 crimes in this district.')
print(f'District with the lowest crime rate: {lowest_crime_district} showing no significantly crimes to registered.')
```

# How did the COVID-19 Pandemic affect overall crime trends in Chicago?

```python
df_ts = df.copy()
df_ts
```

```python
df_ts['datetime'] = pd.to_datetime(df['Date'])
df_ts.info()
```

```python
df_ts = df_ts.set_index('datetime')
df_ts
```

```python
df_ts = df_ts.resample('D').sum()
```

```python
before_pandemic = df_ts['2018':'2019']
during_pandemic = df_ts['2020':'2021']
after_pandemic = df_ts['2022']
```

```python
import matplotlib.pyplot as plt

window_size = 7  
plt.figure(figsize=(12,3))

for period,label in zip([before_pandemic, during_pandemic, after_pandemic],['Before', 'During', 'After']):
    rolling_mean = period.rolling(window=window_size).mean()
    plt.plot(rolling_mean)

plt.xticks(pd.date_range(start='2018', end='2022', freq='Q'), rotation=45)
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(90))

plt.title('Overall Crime Trends Before, During, and After the Pandemic')
plt.xlabel('Year')
plt.ylabel('Average Daily Crimes')
plt.show();
```

- As shown above there was a slight increased in the first quarter of 2020, during the pandemic. Before the pandemic the trend would increase then decrease at small rates, we notice a slight drop in avergae crimes in the last quarter of 2018. After the pandemic, the trend seems to increase with slight depreciatiion in crime. 

# How has the variance in crime changed over the years?

```python
df_var = df.copy()
```

```python
df_var['datetime'] = pd.to_datetime(df_var['Date'])
df_var.info()
```

```python
df_var = df_var.set_index('datetime')
df_ts
```

```python
data_resampled = df_var.resample('Y').sum()
data_resampled
```

```python
var_changes = data_resampled.diff()

var = var_changes.var(axis=1)

var_diff = var.diff()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(var_diff)
plt.title('Changes in Variance of Crime Over the Years')
plt.xlabel('Years')
plt.ylabel('Variance Changes')
plt.grid(True)
plt.show()
```

The variance of crime over the years can be described at first as having a low variance, between the years 2004-2016.The crime dropped had a low dip between 2010-2011. The biggest variance in crime was a huge increase from the year 2020-2021. The variance increased more than 3 points.
